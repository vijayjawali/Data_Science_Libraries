```python
from matplotlib import pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
```

```python
plt.style.use('fivethirtyeight')

slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels= labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [60, 40]
labels = ['Sixty', 'Forty']

plt.pie(slices, labels= labels, wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', 'Extra', 'Extra1']
blue = '#008fd5'
red = '#fc4f30'
yellow = '#e5ae37'
green = '#6d904f'
colors = [blue, yellow, green, red]

plt.pie(slices, labels= labels, colors= colors, wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.2, 0]
```

```python
plt.pie(slices, 
        labels= labels,
        explode = explode,  
        wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, 
        labels= labels,
        explode = explode, 
        shadow= True, 
        wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, 
        labels= labels,
        explode = explode, 
        shadow= True, 
        startangle= 90,
        wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.pie(slices, 
        labels= labels,
        explode = explode, 
        shadow= True, 
        startangle= 90,
        autopct='%1.1f%%',
        wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
